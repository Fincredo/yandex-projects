## Описание проекта

Необходимо подготовить прототип модели машинного обучения для компании «Цифра», которая разрабатывает решения для эффективной работы промышленных предприятий. Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Она поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Для исследования предоставлены данные с параметрами добычи и очистки. Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков. Данные индексируются датой и временем получения информации, при этом соседние по времени параметры часто похожи.

Стадии процесса получения золота из руды:
1. После первичной обработки руды дроблёная смесь отправляется на флотацию (обогащение). После обогащения получается черновой концентрат и «отвальные хвосты» (остатки продукта с низкой концентрацией ценных металлов).
2. Черновой концентрат проходит двухэтапную очистку. На выходе получается финальный концентрат и новые отвальные хвосты.

Эффективность обогащения рассчитывается по формуле: $ Recovery = \frac{C\times(F-T)}{F\times(C-T)} $, где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах (как в финальном продукте, так и в черновом концентрате). Для решения задачи будет использована метрика качества — sMAPE (симметричное среднее абсолютное процентное отклонение), которая вычисляется по формуле:
$ sMAPE = \frac{1}{N} \displaystyle\sum^n_{i=1} \frac{|y_i - \hat{y}_i|}{(|y_i| + |\hat{y}_i|) / 2} \times 100\% $, где:
- y$_i$ — значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество;
- ŷ$_i$ — значение предсказания для объекта с порядковым номером i, например, в тестовой выборке;
- N — количество объектов в выборке.

Нужно спрогнозировать эффективности обогащения как чернового, так и финального концентратов. Итоговая метрика складывается из двух величин: $ \textit {Итоговое} \ sMAPE = 25\% \times sMAPE(rougher) + 75\% \times sMAPE(final) $.

### Проверка эффективности обогащения

Проверим верность расчета эффективности обогащения, вычислив ее на обучающей выборке для признака `'rougher.output.recovery'`.

Формула расчета эффективности обогащения (для этапа флотации):
$$ Recovery = \frac{C\times(F-T)}{F\times(C-T)} $$
где:
- C — доля золота в концентрате после флотации (`'rougher.output.concentrate_au'`);
- F — доля золота в сырье до флотации (`'rougher.input.feed_au'`);
- T — доля золота в отвальных хвостах после флотации (`'rougher.output.tail_au'`).

```python
# Создаем датафрейм с нужными столбцами.
# Просматриваем первые строки и общую информацию о датафрейме.

mae_columns = ['rougher.output.concentrate_au',
               'rougher.input.feed_au',
               'rougher.output.tail_au',
               'rougher.output.recovery']

mae_data = df_train[mae_columns]
display(mae_data.head())

mae_data.info()
```

```python
# Вычисляем эффективность обогащения и находим среднее абсолютное отклонение.

C = mae_data['rougher.output.concentrate_au']
F = mae_data['rougher.input.feed_au']
T = mae_data['rougher.output.tail_au']

mae_data['recovery_calculated'] = (C * (F - T)) / (F * (C - T)) * 100
print('Метрика MAE составила',
      mean_absolute_error(mae_data['rougher.output.recovery'],
                          mae_data['recovery_calculated'])
     )
```

Значение MAE ничтожно мало. Следовательно, эффективность обогащения в таблице была указано верно.

## Модель

### Подготовка

```python
# Составляем функцию для расчета sMAPE.

def smape(target, predictions):
    num = abs(predictions - target)
    den = (abs(target) + abs(predictions)) / 2
    sigma = (num / den).sum()
    smape = 1 / len(target) * sigma * 100
    return smape
```

```python
# Составляем cписок признаков для каждой выборки

features_train = df_train.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_train = df_train['rougher.output.recovery']
target_final_train = df_train['final.output.recovery']

features_test = df_test.drop(['rougher.output.recovery','final.output.recovery'], axis=1)
target_rougher_test = df_test['rougher.output.recovery']
target_final_test = df_test['final.output.recovery']

print('Размеры обучающей выборки с признаками:', features_train.shape)
print('Размеры обучающей выборки с целевым признаком (для этапа флотации):', target_rougher_train.shape)
print('Размеры обучающей выборки с целевым признаком (для финального этапа):', target_final_train.shape)
print()
print('Размеры тестовой выборки с признаками:', features_test.shape)
print('Размеры тестовой выборки с целевым признаком (для этапа флотации):', target_rougher_test.shape)
print('Размеры тестовой выборки с целевым признаком (для финального этапа):', target_final_test.shape)
```

```python
# Масштабируем признаки для этапа флотации, cоздав предварительно их копию.
# Проверяем результаты

features_train_unscaled = features_train.copy()

numeric = features_train.columns
scaler = StandardScaler()
scaler.fit(features_train[numeric])
features_train[numeric] = scaler.transform(features_train[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])

display(features_train_unscaled.head(1))
display(features_train.head(1))
```

```python
# Составляем функцию для кросс-валидации.

def cross_val(model, features, target, smp):
    scorer = make_scorer(smp, greater_is_better=True)
    pipeline = make_pipeline(StandardScaler(), model)
    scores = cross_val_score(pipeline, features, target, scoring=scorer, cv=4)
    score_mean = np.mean(scores)
    return score_mean
```

```python
# Составляем функцию для оценки качества моделей кросс-валидацией (расчет итоговой sMAPE).

def smape_calc(model, features, target_r, target_f, smp):
    smape_rougher = cross_val(model, features, target_r, smp)
    smape_final = cross_val(model, features, target_f, smp)
    smape_total = 0.25 * smape_rougher + 0.75 * smape_final
    print(f'Итоговая sMAPE (модель {model}):')
    return smape_total
```

### Обучение и проверка модели

Обучим разные модели для выбора лучшей.

```python
# Сохраняем дополнительные аргументы для функции smape_calc() в одну переменную.

train_args = features_train_unscaled, target_rougher_train, target_final_train, smape
```

```python
# Модель линейной регрессии

model_lr = LinearRegression()
smape_calc(model_lr, *train_args)
```

```python
# Модель регрессора дерева решений

model_dtr = DecisionTreeRegressor(random_state=12345)
smape_calc(model_dtr, *train_args)
```

```python
%%time
# Модели регрессора случайного леса
# Примерное время обработки - 20 мин 

parameters = {'n_estimators': range (80, 101, 10),'max_depth': range (7, 9)}
grid_scorer = make_scorer(smape, greater_is_better=False)
model_rfr = RandomForestRegressor(random_state=12345)
grid = GridSearchCV(model_rfr, parameters, scoring=grid_scorer, cv=4)

grid_r = grid.fit(features_train_unscaled, target_rougher_train)
print('Для этапа флотации:', grid_r.best_params_)
model_rfr_r = RandomForestRegressor(random_state=12345, **grid_r.best_params_)
smape_rougher = cross_val(model_rfr_r, features_train_unscaled, target_rougher_train, smape)

grid_f = grid.fit(features_train_unscaled, target_final_train)
print('Для финального этапа:',grid_f.best_params_)
model_rfr_f = RandomForestRegressor(random_state=12345, **grid_f.best_params_)
smape_final = cross_val(model_rfr_f, features_train_unscaled, target_final_train, smape)

smape_total = 0.25 * smape_rougher + 0.75 * smape_final
print(f'\nИтоговая sMAPE (модели model_rfr_r и model_rfr_f):', smape_total)
```

```python
# Фиктивная модель

model_dr = DummyRegressor()
smape_calc(model_dr, *train_args)
```

Видим, что случайный лес имеет наилучшую среднюю оценку качества. Поэтому на тестовой выборке проверим модели именно этой структуры. Выясним, в каком из вариантов метрика будет лучше ("model_rfr_r", "model_rfr_f" или их комбинация).

```python
# Функция для проверки моделей "model_rfr" на тестовой выборке

def model_test(model1, model2):
    
    model_rougher = model1
    model_rougher.fit(features_train, target_rougher_train)
    predictions_rougher = model_rougher.predict(features_test)

    model_final = model2
    model_final.fit(features_train, target_final_train)
    predictions_final = model_final.predict(features_test)

    smape_rougher = smape(target_rougher_test, predictions_rougher)
    smape_final = smape(target_final_test, predictions_final)
    smape_total = 0.25 * smape_rougher + 0.75 * smape_final
    print(f'Итоговое sMAPE для тестовой выборки:', smape_total)
```

```python
%%time
# Проверка на тестовой выборке (модели "model_rfr_r" и "model_rfr_f")
# Примерное время обработки - 53 сек

model_test(model_rfr_r, model_rfr_f)
```

```python
%%time
# Проверка на тестовой выборке (только модель "model_rfr_r")
# Примерное время обработки - 50 сек

model_test(model_rfr_r, model_rfr_r)
```

```python
%%time
# Проверка на тестовой выборке (только модель "model_rfr_f")
# Примерное время обработки - 57 сек

model_test(model_rfr_f, model_rfr_f)
```

Таким образом, в случае обоих целевых признаков на тестовой выборке лучшей оказалась модель "model_rfr_f".

Проверим модель на адекватность.

```python
# Проверка на тестовой выборке (модель DummyRegressor)

model_rougher = model_dr
model_rougher.fit(features_train, target_rougher_train)
predictions_rougher = model_rougher.predict(features_test)

model_final = model_dr
model_final.fit(features_train, target_final_train)
predictions_final = model_final.predict(features_test)

smape_rougher = smape(target_rougher_test, predictions_rougher)
smape_final = smape(target_final_test, predictions_final)
smape_total = 0.25 * smape_rougher + 0.75 * smape_final
print(f'Итоговое sMAPE для тестовой выборки (модель {model_dr}):', smape_total)
```

9. __Модель__. В ходе процесса создания, обучения и проверки модели написана функция для расчета sMAPE, подготовлены обучающие и целевые признаки (обучающие признаки отмасштабированы). Также составлены функции для кросс-валидации и поиска с ее помощью модели с наилучшим итоговым sMAPE. В сравнении разных моделей с использованием методов кросс-валидации лучший результат показала модель регрессора случайного леса с количеством деревьев 90 и глубиной 8. Финальная sMAPE для данной модели на тестовой выборке составила 7.19 (больше, чем результаты фиктивной модели).

Добыча и обработка золотносной руды - один из тех процессов, где каждая деталь в параметрах может иметь колоссальное влияние на готовый продукти и, соответственно, на прибыль предприятия. Правильный прогноз такого параметра, как эффективность обогащения руды, позволит подбирать такие соотношения исходных данных, которые максимизируют процент содержания золота в конечном концентрате. Это позволит сэкономить средства на различные реальные эксперименты с подобными соотношениями.

Сложность в таких исследованиях чаще всего заключается в неправильном истолковании причин неполноты исходных данных, а также аномальных значений в них. Порой не удаленные вовремя из датасета наблюдения или, наоборот, - преждевременно удаленные, - могут вызвать сильные девиации в предсказаниях модели. Поэтому важное значение для исследования имеет правильная интерпретация и предобработка данных. В частности, на мой взгляд, были необходимы следующие шаги:

1. Анализ отсутствующих в тестовой выборке признаков. Прежде всего здесь важную роль сыграл сам факт предоставления этой выборки без лишних признаков, из чего можно было уже сделать вывод об их очевидной недоступности при прогнозе в реальных условиях. Это в свою очередь помогло проигнорировать аномалии, имевшиеся в лишних признаках.


2. Проверка правильности расчета в исходных данных эффективности обогащения руды. Неверный хоть на долю процента коэффициент мог бы сильно исказить все последующие звенья расчетов.


3. Выбор оптимального метода заполнения пропусков в данных. В случае, когда доля пропусков в отдельных признаках доходит почти до 10%, очень важно не ошибиться с методом. И здесь любая исходная информации в техническом задании может сыграть ключевую роль, в частности, информация о схожести данных в идущих подряд наблюдениях.


4. Выбор последовательности заполнения пропусков. Например, то, что заполнены сначала пропуски в исходном датасете, а затем результаты перенесены в обучающую и тестовую выборки. Здесь главную роль сыграло наличие уникального идентификатора - даты и времени наблюдения.


5. Своевременное удаление наблюдений с нулевыми значениями в целевых признаках. Важно при этом видеть распределение и понимать, что таких значений аномальное количество, и их наличие может снизить качество будущей модели.


6. Удаление лишних признаков из обучающей выборки. Невыполнение данного шага могло привести, например, к переобучению модели, что снизило бы ее эффективность при предсказаниях на тестовой выборке.


7. Анализ изменения концентрации отдельных металлов, а также их суммарной концентрации, на разных этапах обогащения. Это позволяет лучше понять технологический процесс, роль каждого металла в нем и важность каждого из его этапов.


8. Сравнение распределений размеров гранул в обучающей и тестовой выборках. При подготовке к обучению модели важность данного фактора - репрезентативности выборки - сложно переоценить.


9. Правильная стандартизация данных. Необходимо как на этапах кросс-валидации, так и на этапе проверки на тестовой выборке соблюсти порядок: масштабировать обучающую выборку, после чего применять параметры масштабирования на валидационной/тестовой.


10. Наконец, правильный подбор гиперпараметров для модели. Здесь главное - не переусердствовать и выбрать оптимальное сочетание всех доступных инструментов для прогнозирования и технических возможностей конкретной машины, на которой производится данный процесс.

Во время исследования сложность представил момент отбора тренировочных признаков для предсказания отдельно каждого из двух целевых (соответственно, для этапа флотации и финального этапа). В итоге было решено использовать в обоих случая одни и те же признаки.

В качестве меры дополнительной оптимизации можно было бы попробовать включить в исходные данные также промежуточную (между 1-ым и 2-ым этапами очистки) эффективность обогащения, которая так же учитывалась бы при обучении модели.

Стоявшая в начале исследования задача выполнена. Найдена модель, делающая прогноз эффективности обогащения руды с минимальной ошибкой.
