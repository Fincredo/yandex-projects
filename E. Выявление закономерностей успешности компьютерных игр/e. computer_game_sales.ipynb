#### Столбец "year_of_release"

У 269 игр не указан год выпуска. Проще всего удалить эти строки или заполнить медианными значениями: это меньше 2% данных. Однако, во-первых, "данных много не бывает". А во-вторых, ради интереса попытаемся восстановить данные, проникнув в секреты, которые они скрывают, найдя неявные закономерности. 

```python
# Просматриваем строки с пропусками в столбце "year_of_release".

games[games['year_of_release'].isna()]
```

Обратим внимание, что в названиях многих игр присутствует год. Детальный просмотр показывает, что он указан в одном из следующих видов: 
- YYYY, 
- YY-YY, 
- 2KY, 
- 2KYY,
- 9Y.

Причем, иногда этот год совпадает с годом выпуска игры, а иногда - нет. 

Найдем все игры с имеющимся годом выпуска, удовлетворяющие этим условиям, и выделим из названий год.

```python
# С помощью регулярных выражений выводим все найденные строки в виде:
# название игры / выделенный из названия год / реальный год выпуска игры.

n = 0
delta = []
for name, year in zip(games['name'], games['year_of_release']):
    rex = re.search(r'\d{2}-?\d{2}|\dK\d?\d|9\d', name)
    if rex is not None:
        try:
            if int(rex.group()) < 100:
                year_found = int(rex.group()) + 1900
            else:
                year_found = int(rex.group())
        except:
            if '-' in rex.group():
                year_found = 2000 + int(rex.group().split('-')[1])
            elif 'K' in rex.group():
                year_found = 2000 + int(rex.group().split('K')[1])
            else:
                print('Не получилось извлечь год')
        if abs(year_found - year) <= 10:
            delta.append(year_found - year)
            print(f"{name} / {year_found} / {year}")
        n += 1
```

```python
# Выводим количество удовлетворяющих условию строк.
# Выводим количество тех из них, 
# где год меньше всего отличается от реального года выпуска игры,
# а также уникальные значения отличий от года выпуска 
# (найденный в названии год минус год выпуска игры).

print('Количество найденных строк с наличием года '
      '(или подобия года) в названии игры:', n
     )
print('Количество игр, где найденный в названии год '
      'отличается от года выпуска не более, чем на 10 лет:', len(delta)
     )
print()
print(pd.Series(delta).value_counts())
```

Как видим, чаще всего встречается различие в один год. Т.е. в большинстве случаев реальный год выпуска на единицу меньше, чем указанный в названии игры.

Теперь найдем аналогичные данные для строк с пропусками в значениях года выпуска.

```python
# С помощью регулярных выражений выводим все найденные строки в виде: 
# название игры / выделенный из названия год / реальный год выпуска игры ("NaN").

n = 0
for name, year in zip(games['name'], games['year_of_release']):
    rex = re.search(r'\d{2}-?\d{2}|\dK\d?\d|9\d', name)
    if np.isnan(year) and rex is not None:
        try:         
            if int(rex.group()) < 100:
                year_found = int(rex.group()) + 1900
            else:
                year_found = int(rex.group())
        except:
            if '-' in rex.group():
                year_found = 2000 + int(rex.group().split('-')[1])
            elif 'K' in rex.group():
                year_found = 2000 + int(rex.group().split('K')[1])
            else:
                print('Не получилось извлечь год')
        print(f"{name} / {year_found} / {year}")
        n += 1
```

```python
# Выводим количество удовлетворяющих условию строк.

print('Количество найденных строк с наличием года в названии игры:', n)
```

Создадим функцию, которая заполнит в найденных строках пропуски значением года из названия, уменьшенным на единицу.

```python
# Функция заполняет пропуски в годах выпуска игр
# соответствующими значениями из названий, уменьшенными на 1.

def find_year(row):
    rex = re.search(r'\d{2}-?\d{2}|\dK\d?\d|9\d', row['name'])
    if pd.isna(row['year_of_release']) and rex is not None:
        try:
            if int(rex.group()) < 100:
                year_found = int(rex.group()) + 1900
            else:
                year_found = int(rex.group())
        except:
            if '-' in rex.group():
                year_found = 2000 + int(rex.group().split('-')[1])
            elif 'K' in rex.group():
                year_found = 2000 + int(rex.group().split('K')[1])
            else:
                print('Не получилось извлечь год')
        return year_found - 1
    return row['year_of_release']
```

```python
# Применяем функцию find_year() к столбцу "year_of_release".
# Проверяем количество оставшихся пропусков в столбце с годом выпуска игр.

games['year_of_release'] = games.apply(find_year, axis=1)
print('Количество пропущенных значений в столбце "year_of_release":',
      games['year_of_release'].isna().sum()
     )
```

Поскольку одна и та же игра выходит на различных платформах примерно в одинаковые годы, часть из оставшихся пропусков заменим на медианные значения по сгруппированным данным (по названию игры).

```python
# Заполняем пустые ячейки столбца с годом выпуска медианными значениями по каждой игре.
# Выводим обновленную информацию о таблице.

games_temp = games.groupby('name')['year_of_release'].transform('median')
games['year_of_release'] = games['year_of_release'].fillna(games_temp)
print('Количество пропущенных значений в столбце "year_of_release":',
      games['year_of_release'].isna().sum()
     )
```

Выведем список уникальных названий игр с пропусками в ячейках с годом выпуска.

```python
names_nan_list = games[games['year_of_release'].isna()]['name'].value_counts().index
print(names_nan_list)
```

Часть игр имеют серии с похожими названиями. Серии чаще всего выходят с разницей в год-два.

Создадим функцию, с помощью которой каждой игре в таблице сопоставим наиболее похожее название из вышеприведенного списка. 

```python
# Функция сравнивает значение в ячейке таблицы с каждым названием в "names_nan_list"
# и возвращает название с максимальным уровнем совпадения, либо - пустое значение 
# (если совпадение меньше минимального). 

def name_match(row):
    match_min = 0.7 # Установленный минимальный уровень совпадения названий
    game = ''
    for name in names_nan_list:
        rex = re.match(r'\w+', name)
        if re.match(rex.group(), row) is not None:
            match_ratio = difflib.SequenceMatcher(None, name, row).ratio()
            if match_ratio > match_min:
                match_min = match_ratio
                game = name
    return game
```

```python
# Применяем функцию name_match() к столбцу "name" и сохраняем в новом столбце "names_nan".

games['names_nan'] = games['name'].apply(name_match)
print("Функция успешно применена.")
```

### Влияние отзывов на глобальные продажи на разных платформах

Для начала составим диаграмму рассеяния и вычислим корреляции между отзывами и продажами для одной платформы. 

```python
# Диаграмма "Scatterplot" для определения корреляции оценок и объемов продаж игр на примере платформы PS4
    
s = games_actual.query("platform == 'PS4'")['total_sales']
c = games_actual.query("platform == 'PS4'")['critic_score']
u = games_actual.query("platform == 'PS4'")['user_score']

f, ax = plt.subplots(1, 2, figsize=(15, 5))
f.suptitle('Корреляция между отзывами и продажами игр на примере платформы "PS4"', size=14)

ax[0].scatter(c, s, c='r')
ax[0].set_title('Данные по отзывам критиков')
ax[0].set(xlabel='Оценки', ylabel='Объемы продаж, млн шт.')
ax[0].set_xticks(range(0, 105, 5))

ax[1].scatter(u, s, c='y')
ax[1].set_title('Данные по отзывам пользователей')  
ax[1].set(xlabel='Оценки', ylabel='Объемы продаж, млн шт.')
ax[1].set_xticks(range(0, 11, 1))

plt.show()

print(f'Коэффициент корреляции между оценками критиков'
      f'и продажами отдельных игр на платформе "PS4": {s.corr(c):.2f}')
print(f'Коэффициент корреляции между оценками пользователей'
      f'и продажами отдельных игр на платформе "PS4": {s.corr(u):.2f}')
```

## Проверка гипотез

### Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем нулевую и альтернативную гипотезы:
    

H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H1: Средние пользовательские рейтинги платформ Xbox One и PC разные.
    

Для проверки гипотез применим t-тест. Сравним две генеральные совокупности (средние значения оценок пользователей по двум разным платформам) через имеющиеся выборки.

```python
# Выводим статистические данные двух выборок
   
x1_u_score_size = games_actual.query('platform == "XOne"')['user_score'].count()
pc_u_score_size = games_actual.query('platform == "PC"')['user_score'].count()
print(f'Размер первой выборки (данные по платформе "Xbox One") - {x1_u_score_size}')
print(f'Размер второй выборки (данные по платформе "PC") - {pc_u_score_size}\n')

x1_u_score_mean = games_actual.query('platform == "XOne"')['user_score'].mean()
pc_u_score_mean = games_actual.query('platform == "PC"')['user_score'].mean()
print(f'Среднее значение в выборке по платформе "Xbox One" - {x1_u_score_mean:.2f}')
print(f'Среднее значение в выборке по платформе "PC" - {pc_u_score_mean:.2f}\n')

x1_u_score_var = games_actual.query('platform == "XOne"')['user_score'].var()
pc_u_score_var = games_actual.query('platform == "PC"')['user_score'].var()
print(f'Дисперсия в выборке по платформе "Xbox One" - {x1_u_score_var:.4f}')
print(f'Дисперсия в выборке по платформе "PC" - {pc_u_score_var:.4f}')
```

Возьмем критический уровень статистической значимости равный 0,03. Т.е. допустимая для нас минимальная вероятность получить такие же или более экстремальные различия в значениях статистик по двум платформам при верности нулевой гипотезы - 3%.


Для выбора параметра equal_var необходимо проверить три критерия:

- генеральные совокупности независимы (+);
- размер каждой выборки (по условию Центральной предельной теоремы) составляет не менее нескольких десятков значений (+);
- дисперсии рассматриваемых генеральных совокупностей равны (-).

Поскольку дисперсии в выборках отличаются, хотя сами выборки примерно равного размера, для параметра equal_var установим значение False.

```python
# Применение t-теста для проверки Гипотезы 1

alpha = 0.03
results = st.ttest_ind(games_actual.query('platform == "XOne" and user_score.notna()')['user_score'], 
                       games_actual.query('platform == "PC" and user_score.notna()')['user_score'],
                       equal_var = False
                      )
print('Уровень значимости (p-value):', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print('Не получилось отвергнуть нулевую гипотезу.')
```

_Вывод_

Гипотеза 1 вполне может быть верной: 

мы не можем утвержать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются.

### Гипотеза 2: Средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем нулевую и альтернативную гипотезы:
    

H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые.

H1: Средние пользовательские рейтинги жанров Action и Sports разные.
    

Для проверки гипотез применим t-тест. Сравним две генеральные совокупности (средние значения оценок пользователей по двум разным жанрам) через имеющиеся выборки.

```python
# Выводим статистические данные двух выборок

action_u_score_size = games_actual.query('genre == "Action"')['user_score'].count()
sports_u_score_size = games_actual.query('genre == "Sports"')['user_score'].count()
print(f'Размер первой выборки (данные по жанру "Action") - {action_u_score_size}')
print(f'Размер второй выборки (данные по жанру "Sports") - {sports_u_score_size}\n')

action_u_score_mean = games_actual.query('genre == "Action"')['user_score'].mean()
sports_u_score_mean = games_actual.query('genre == "Sports"')['user_score'].mean()
print(f'Среднее значение в выборке по платформе "Action" - {action_u_score_mean:.2f}')
print(f'Среднее значение в выборке по платформе "Sports" - {sports_u_score_mean:.2f}\n')

action_u_score_var = games_actual.query('genre == "Action"')['user_score'].var()
sports_u_score_var = games_actual.query('genre == "Sports"')['user_score'].var()
print(f'Дисперсия в выборке по платформе "Action" - {x1_u_score_var:.4f}')
print(f'Дисперсия в выборке по платформе "Sports" - {pc_u_score_var:.4f}')
```

Возьмем критический уровень статистической значимости равный 0,01. Т.е. допустимая для нас минимальная вероятность получить такие же или более экстремальные различия в значениях статистик по двум жанрам при верности нулевой гипотезы - 1%.


Для выбора параметра equal_var необходимо проверить три критерия:

- генеральные совокупности независимы (+);
- размер каждой выборки (по условию Центральной предельной теоремы) составляет не менее нескольких десятков значений (+);
- дисперсии рассматриваемых генеральных совокупностей равны (-).

Поскольку дисперсии в выборках отличаются, а сами выборки разного размера, для параметра equal_var установим значение False.

```python
# Применение t-теста для проверки Гипотезы 2

alpha = 0.01
results = st.ttest_ind(games_actual.query('genre == "Action" and user_score.notna()')['user_score'], 
                       games_actual.query('genre == "Sports" and user_score.notna()')['user_score'],
                       equal_var = False
                      )
print('Уровень значимости (p-value):', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print('Не получилось отвергнуть нулевую гипотезу.')
```
