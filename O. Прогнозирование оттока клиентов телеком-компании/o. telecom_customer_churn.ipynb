## Описание

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Для исследования предоставлены персональные данные о некоторых клиентах, информация об их тарифах и договорах. Сведения актуальны на 1 февраля 2020.

Оператор предоставляет два основных типа услуг: 
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию или оптоволоконный кабель.  

Также доступны дополнительные услуги: Интернет-безопасность: антивирус и блокировка небезопасных сайтов; выделенная линия технической поддержки; облачное хранилище файлов для резервного копирования данных; стриминговое телевидение и каталог фильмов.

Клиенты могут оплачивать услуги каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Выводы

Для исследования предоставлено 4 таблицы: `contract`, `personal`, `internet` и `phone`. В ходе предварительного изучения данных выявлено следующее:

1. Таблица `contract`.

    1.1. Датафрейм состоит из 7043 строк и следующих 8 столбцов:
        customerID - идентификатор абонента;
        BeginDate - дата заключения договора обслуживания;
        EndDate	- дата расторжения договора обслуживания;
        Type - тип договора (по периодичности платежей);
        PaperlessBilling - способ получения счета на оплату;
        PaymentMethod - способ платежа;
        MonthlyCharges - среднемесячный платеж;
        TotalCharges - суммарный платеж.
        
    1.2. Тип данных во всех столбцах, кроме 'MonthlyCharges', - *object*. У последнего - *float*.
    
    1.3. Столбец 'customerID' содержит коды из 4 цифр, дефиса и 5 заглавных букв, и присутствует во всех четырех таблицах. Данный столбец содержит 7043 уникальных идентификатора, что указывает на то, что каждому абоненту соответствует один договор.
    
    1.4. Столбцы 'BeginDate' и 'EndDate' содержат, соответственно, данные о дате и дате со временем (либо пометку "No", если договор не расторгнут). Их тип целесообразно сменить на *datetime*.
    
    1.5. Столбцы 'Type', 'PaperlessBilling' и 'PaymentMethod' содержат категориальные признаки. В столбце 'Type' имеется 3 уникальных значения: помесячная (55% случаев), годичная (24%) и двухгодичная (21%) оплата. Значения в 'PaperlessBilling' (получение счета в электронном виде) представлены в 2 вариантах: да (59.2%) и нет (40.8%). Способов оплаты ('PaymentMethod') всего 4: электронным чеком (33.6%), почтовым чеком (22.9%), списанием с банковского счета (21.9%), списанием с кредитной карты (21.6%).
    
    1.6. Столбцы 'MonthlyCharges' и 'TotalCharges' содержат количественные признаки.
    
    1.7. Без учета столбца с ID, в таблице имеется 45 дубликатов, или около 0.6%, что вполне можно списать на простое совпадение.
    
    1.8. В статистиках количественных признаков (а конкретно, данных столбца 'MonthlyCharges') аномалий нет, все характеристики находятся в пределах нормы. Для исследования статистик столбца 'TotalCharges' тип его данных необходимо сначала сменить на *float*.
    
    1.9. Распределение среднемесячных платежей имеет две вершины: высокую - в районе 20 у.е. (1200 случаев, или 17%) и умеренную - в районе 80 у.е. ## Предобработка данных

### Объединение таблиц и переименование столбцов

Объединим все четыре таблицы в одну по столбцу `customerID`, сменим регистр в названиях столбцов, а также удалим общий столбец с идентификатором абонента, сделав его значения индексами строк.

```python
# Объединяем таблицы.

df = (contract
      .merge(personal, how='left', on='customerID')
      .merge(internet, how='left', on='customerID')
      .merge(phone, how='left', on='customerID'))
```

```python
# Переводим названия столбцов из "верблюжьего" регистра в "змеиный".
# Просматриваем обновленный список индексов.

df.columns = df.columns.str.replace(r"([a-z])([A-Z])", r"\1_\2").str.lower()

df.columns
```

```python
# Устанавливаем в качестве индексов значения столбца 'customer_id'.
# Просматриваем первые строки обновленного датафрейма.

df = df.set_index('customer_id', drop=True)
df.head()
```

### Добавление столбцов и изменение типов данных

Для целей исследования нам необходимо выделить целевой признак. Для этого создадим новый столбец, значения в котором будут составлять единицы и нули, соответствующие наличию или отсутствию факта ухода абонента. Данные для этого возьмем из столбца `end_date`.

```python
# Добавляем столбец 'exited' с целевым признаком (факт ухода абонента).

df['exited'] = (df['end_date'] != 'No').astype('int')
```

Сменим тип данных в столбцах с датами заключения и расторжения договора на *datetime*. Значения "No" в столбце `end_date` заменим на последнюю дату - "2020-02-01". Добавим столбец с разницей между двумя датами - количеством дней обслуживания абонента.

```python
# Меняем тип данных столбцов 'begin_date' и 'end_date' на datetime.
# В столбце 'end_date' заменяем 'No' на дату '2020-02-01'.

df['begin_date'] = pd.to_datetime(df['begin_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'].replace({'No': '2020-02-01'}), errors='coerce')
```

```python
# Добавляем столбец 'days' с количеством дней обслуживания абонента.

df.insert(2, 'days', (df['end_date'] - df['begin_date']))
df['days'] = df['days'].dt.days
```

Для отображения статистик по столбцу `total_charges` и возможности дальнейшей его обработки сменим тип данных в нем на числовой.

```python
# Меняем тип данных столбца 'total_charges' на числовой.

df['total_charges'] = pd.to_numeric(df['total_charges'], errors='coerce')
```

```python
# Просматриваем первые строки обновленного датафрейма.

df.head()
```

```python
# Просматриваем общую информацию об обновленном датафрейме.

df.info()
```

### Обработка пропусков

В столбце с суммарными платежами появились пропуски. Просмотрим соответствующие строки.

```python
# Проверяем строки с образовавшимися пропусками в столбце 'total_charges'.

df[df['total_charges'].isna()]
```

Как видим, все эти 11 договоров были заключены 01.02.2020 г., т.е. в дату формирования базы. Соответственно, данные абоненты еще не успели воспользоваться услугами компании. Поэтому эти строки будет правильным удалить.

```python
# Удаляем строки с пропусками в столбце 'total_charges'.
# Обновляем данные о пропусках.

df.dropna(subset=['total_charges'], inplace=True)
df.isna().sum()
```

Что касается пропусков в столбце с данными об услугах телефонии и 7 столбцах с данными об услугах, касающихся интернета, то необходимо сначала понять их природу.

Сперва проверим, есть ли зависимость между наличием пропусков в столбце `multiple_lines` и датой заключения договоров обслуживания. Например, выясним, не связано ли наличие пропусков с тем, что соответствующие договоры были заключены давно, когда еще не стоял выбор между одной или несколькими телефонными линиями (и в этом случае можно было бы просто заменить пропуски на значение "No").

```python
# Гистограммы распределения дат начала обслуживания 
# для строк с пропусками в столбце 'multiple_lines' и без пропусков

plt.hist(df[df['multiple_lines'].notna()]['begin_date']) # без пропусков (синий график)
plt.hist(df[df['multiple_lines'].isna()]['begin_date']); # с пропусками (оранжевый график)
```

Из графика явствует, что такие договоры заключались в разное время, следовательно значения этого параметра ('begin_date') не связаны с наличием пропусков в столбце `multiple_lines`.

Теперь аналогично проверим столбцы `internet_service`, `online_security`, `online_backup`, `device_protection`, `tech_support`, `streaming_tv`, `streaming_movies`.

```python
# Наложенные друг на друга гистограммы распределения дат начала обслуживания 
# для строк с пропусками в 7 столбцах об интернет-услугах и без пропусков

i_services = ['internet_service',
               'online_security',
               'online_backup',
               'device_protection',
               'tech_support',
               'streaming_tv',
               'streaming_movies']

for col in i_services:
    plt.hist(df[df[col].notna()]['begin_date']) # без пропусков (зеленый график)
    plt.hist(df[df[col].isna()]['begin_date']) # с пропусками (красный график)
```

Из графиков следует, что:
- во-первых, значения параметра 'begin_date' для этих 7 столбцов также не связаны с наличием в них пропусков;
- во-вторых, совпадение наложенных друг на друга 7 графиков означает совпадение строк с пропусками во всех 7 столбцах (проверим это).

```python
# Проверяем количество случаев, когда в столбце 'internet_service' имеется пропуск
# и при этом в каком-либо из остальных 6 столбцов бывшей таблицы "internet" пропуска нет.

difference = 0
for col in i_services[1:]:
    difference += df[(df['internet_service'].isna())
                     &((df[col].notna()))]['begin_date'].count()
print("Общее количество несовпадений:", difference)
```

Исходя из того, что все интернет-услуги либо одновременно имеют пропуски, либо одновременно - не имеют, можно сделать вывод, что наличие пропусков свидетельствует о полном отсутствии интернет-услуг у соответствующего абонента. В этом случае, во всех 6 столбцах, кроме `internet_service`, можно заменить пропуски на "No".

```python
# Заполняем значением "No" пропуски в столбцах с допуслугами, связанными с интернетом.

for col in i_services[1:]:
    df[col] = df[col].fillna('No')
```

Проверим, как соотносится наличие пропусков в таблицах `internet_service` и `multiple_lines`.

```python
# Выявление количества случаев пользования каждой из услуг по отдельности и их комбинацией

both = df[df['internet_service'].notna() 
          & df['multiple_lines'].notna()]['begin_date'].count()
only_phone = df[df['internet_service'].isna() 
                & df['multiple_lines'].notna()]['begin_date'].count()
only_internet = df[df['internet_service'].notna() 
                   & df['multiple_lines'].isna()]['begin_date'].count()
none = df[df['internet_service'].isna() 
          & df['multiple_lines'].isna()]['begin_date'].count()

print("Количество случаев, "
      "когда пропусков нет в обоих столбцах:", both)
print("Количество случаев, "
      "когда пропуск только в столбце 'internet_service':", only_phone)
print("Количество случаев, "
      "когда пропуск только в столбце 'multiple_lines':", only_internet)
print("Количество случаев, "
      "когда пропуски есть в обоих столбцах:", none)
print("Суммарное количество во всех четырех вариантах:", 
      both + only_phone + only_internet + none)
```

Наиболее очевидный вывод: пропуски означают полное отсутствие подключения соответствующей услуги. У 680 абонентов нет телефонной линии, а у 1520 других абонентов нет интернет-услуг. При этом одновременно обеих услуг у абонента не может не быть, что вполне логично, и означает, что договор обслуживания предполагает наличие хотя бы одной из них.

Таким образом, пропуски в этих двух столбцах необходимо заполнить нейтральными значениями. Например, для интернета - "No internet", а для телефонии - "No phone".

```python
# Заполняем пропуски в столбцах 'internet_service' и 'multiple_lines'.

df['internet_service'] = df['internet_service'].fillna('No internet')
df['multiple_lines'] = df['multiple_lines'].fillna('No phone')
```

### Кодирование значений категориальных признаков

Перед обучением моделей применим к категориальным признакам кодирование методом OHE.

```python
# Выделяем категориальные признаки 
# и производим кодирование с помощью OneHotEncoder.

cat_features = ['type', 'payment_method', 'gender', 
                'internet_service', 'multiple_lines']
enc = OneHotEncoder()
enc = enc.fit(features_train[cat_features])
enc_columns = np.concatenate(enc.categories_)
```

Применим кодировку на обеих выборках (train и test). Во избежание дублирования кода, напишем соответствующую функцию.

```python
# Функция для кодирования категориальных признаков методом OHE

def encoder(f):
    enc_arr = enc.transform(f[cat_features]).toarray()
    enc_df = f.join(pd.DataFrame(enc_arr, 
                                 columns=enc_columns, 
                                 index=f.index).astype('int'))
    
    enc_df.drop(['type', 'payment_method', 'gender', 
                 'internet_service', 'multiple_lines', 
                 'Month-to-month', 'Bank transfer (automatic)', 
                 'Female', 'No internet', 'No phone'], 
                axis=1, inplace=True)

    enc_df.rename({'One year': 'type_1_year', 'Two year': 'type_2_year',
                   'Credit card (automatic)': 'credit_card',
                   'Electronic check': 'electronic_check',
                   'Mailed check': 'mailed_check',
                   'Male': 'is_male',
                   'DSL': 'dsl',
                   'Fiber optic': 'fiber_optic',
                   'No': 'single_line',
                   'Yes': 'multiple_lines'}, 
                  axis=1, inplace=True)
    
    enc_df.replace({'No': 0, 'Yes': 1}, inplace=True)
    
    cols = list(enc_df)
    cols.insert(3, cols.pop(cols.index('is_male')))
    enc_df = enc_df.loc[:, cols]

    return enc_df
```

```python
# Запускаем функцию для каждой выборки. 

try:
    features_train = encoder(features_train)
    features_test = encoder(features_test)
    
except:
    print('Возможно, Вы уже проделали эту операцию.')
```

### Масштабирование признаков и борьба с дисбалансом классов

```python
# Стандартизируем количественные признаки обучающей выборки (путем масштабирования), 
# cоздав предварительно ее копию.

features_train_unscaled = features_train.copy()
numeric = ['days', 'monthly_charges']
scaler = StandardScaler()
scaler = scaler.fit(features_train[numeric])
```

Составим функцию для кросс-валидации и вывода метрик (AUC-ROC, Accuracy, Precision, Recall, F1-score). В процессе кросс-валидации используем немасштабированную обучающую выборку и пайплайн во избежание переобучения валидационных выборок.

```python
# Составляем функцию для вывода метрик через кросс-валидацию.

def metrics_cv(model):
    score_dict = {}
    score_lst = ['roc_auc', 'accuracy', 'precision', 'recall', 'f1']
    for s in score_lst:
        pipeline = make_pipeline(StandardScaler(), model)
        scores = cross_val_score(pipeline, features_train_unscaled, target_train, 
                                 scoring=s, cv=4)
        score_dict[s] = np.mean(scores)
        print(f'Значение {s}:', '{:.4f}'.format(score_dict[s]))
```

Т.е. в каждом наборе одна из них будет иметь свое максимально возможное значение в установленных границах порога вероятности (эмпирическим способом подобраны оптимальные границы от 0.2 до 0.5, с шагом 0.005). Помимо указанных метрик также выведем матрицы ошибок для каждого случая. Для сокращения кода создадим соответствующую функцию. 

```python
# Функция для поиска лучшего порога для каждой метрики и вывода метрик:
# accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

def metrics(score):
    scr_max = 0
    best_threshold = 0
    best_predictions = 0
    for threshold in np.arange(0.2, 0.5, 0.005):
        predictions = probabilities[:, 1] > threshold
        scr = score(target_test, predictions)
        if scr > scr_max:
            scr_max = scr
            best_threshold = threshold
            best_predictions = predictions

    print(f'Лучший порог для {score.__name__}: {best_threshold:.3f}')
 
    accuracy = accuracy_score(target_test, best_predictions)
    precision = precision_score(target_test, best_predictions)
    recall = recall_score(target_test, best_predictions)
    f1 = f1_score(target_test, best_predictions)
    matrix = confusion_matrix(target_test, best_predictions) 

    print('\nМЕТРИКИ')
    print('Правильность (Accuracy):', '{:.4f}'.format(accuracy))
    print('Точность (Precision):', '{:.4f}'.format(precision))
    print('Полнота (Recall):', '{:.4f}'.format(recall))
    print('F1-мера (F1-score):', '{:.4f}'.format(f1))
    print('\nМАТРИЦА ОШИБОК')
    print(f'Истинно-отрицательные ответы (TN): {matrix[0, 0]} '
          f'({matrix[0, 0] / (matrix[0, 0] + matrix[0, 1]):.1%} отрицательного класса)')
    print(f'Ложно-отрицательные ответы (FN): {matrix[1, 0]} '
          f'({matrix[1, 0] / (matrix[1, 0] + matrix[1, 1]):.1%} положительного класса)')
    print(f'Истинно-положительные ответы (TP): {matrix[1, 1]} '
          f'({matrix[1, 1] / (matrix[1, 0] + matrix[1, 1]):.1%} положительного класса)')
    print(f'Ложно-положительные ответы (FP): {matrix[0, 1]} '
          f'({matrix[0, 1] / (matrix[0, 0] + matrix[0, 1]):.1%} отрицательного класса)')
```

### Дополнительные исследования

Итак, у нас есть 4 стратегии для оператора связи: 
- максимизация Accuracy;
- максимизация Precision;
- максимизация Recall; 
- максимизация F1-меры.

Рассмотрим каждый вариант по отдельности. Начнем с Precision и Recall.

1. Precision - метрика, показывающая, какая доля из выбранных моделью абонентов действительно могла уйти. Это важно, если компания беспокоится о своих расходах, которые понадобятся для удержания этих абонентов.

    $ Presicion = TP / (TP + FP) $


2. Recall - метрика, показывающая, какую долю из действительно собиравшихся уйти абонентов выбрала модель. Это имеет значение, если компания больше беспокоится об удержании абонентов, чем о связанных с этим расходах. 

    $ Recall = TP / (TP + FN) $


3. F1-мера сочетает в себе оба вышеуказанных варианта. Являясь средней гармонической между точностью и полнотой,  F1-мера представляет собой их гармоничное сочетание. 

    $ F1 = TP / (TP + F/2) $


4. Accuracy, в отличие от предыдущих метрик, просто указывает долю правильных предсказаний, данных моделью, независимо от предсказываемых классов.

    $ Accuracy = T / (T + F) $


Остается предложить компании выбрать приемлемую для себя стратегию. Для этого сравним полученные цифры. 


**1.** При максимальной **точности** (0.8806) матрица ошибок была следующей:
- Истинно-отрицательные ответы (TN): 1245;
- Ложно-отрицательные ответы (FN): 136;
- Истинно-положительные ответы (TP): 332;
- Ложно-положительные ответы (FP): 45.

Компания выберет (332 + 45) потенциально "ненадежных" абонентов, чтобы их удержать. При этом еще 136 реально "ненадежных" абонентов она не заметит и упустит.

Предположим, компания может в дальнейшем получить чистый доход от одного удержанного абонента, равный $x$. А затраты, которые она понесет для его удержания, составят $y$. Соответственно, прибыль будет равна $ x-y $. Тогда при этой стратегии общая прибыль оператора будет равна:

$ 332x - (332 + 45)y - 136(x - y) $, 

где $136(x-y)$ - это недополученная прибыль, т.е., фактически, убыток.

Упрощаем выражение:

$ 332x - (332 + 45)y - 136(x - y) = 332x - 332y - 45y - 136x + 136y = 196x - 241y $.

Таким образом, для того, чтобы это было выгодно компании, $196x$ должно быть больше, чем $241y$, или $x > 1.23y$.

Т.е. компания может пойти по этой стратегии, если ее потенциальный доход от каждого абонента превысит расход на его удержание более чем в 1.23 раз, или на 23%. При этом оператор должен быть готов потерять 136 абонентов.


Проведем те же расчеты для остальных стратегий.


**2.** При максимальной **полноте** (0.8697) матрица ошибок была следующей:
- Истинно-отрицательные ответы (TN): 1090;
- Ложно-отрицательные ответы (FN): 61;
- Истинно-положительные ответы (TP): 407;
- Ложно-положительные ответы (FP): 200.

$ 407x - (407 + 200)y - 61(x - y) = 407x - 407y - 200y - 61x + 61y = 346x - 546y $.

Таким образом, для того, чтобы это было выгодно компании, $346x$ должно быть больше, чем $546y$, или $x > 1.58y$.

Т.е. компания может использовать эту стратегию, если ее потенциальный доход от абонента превысит расход более чем в 1.58 раз, или на 58%. При этом от нее уйдет 61 абонент.


**3.** При максимальной **F1-мере** (0.7914) матрица ошибок была следующей:
- Истинно-отрицательные ответы (TN): 1222;
- Ложно-отрицательные ответы (FN): 117;
- Истинно-положительные ответы (TP): 351;
- Ложно-положительные ответы (FP): 68.

$ 351x - (351 + 68)y - 117(x - y) = 351x - 351y - 68y - 117x + 117y = 234x - 302y $.

Таким образом, для того, чтобы это было выгодно компании, $234x$ должно быть больше, чем $302y$, или $x > 1.29y$.

Т.е. компания может пойти по этому пути, если доход от абонента может превысить расход более чем в 1.29 раз, или на 29%. Но при этом она потеряет по недосмотру 117 абонентов.


**4.** При максимальной **правильности** (0.8976) матрица ошибок была следующей:
- Истинно-отрицательные ответы (TN): 1243;
- Ложно-отрицательные ответы (FN): 133;
- Истинно-положительные ответы (TP): 335;
- Ложно-положительные ответы (FP): 47.

$ 335x - (335 + 47)y - 133(x - y) = 335x -335y - 47y - 133x + 133y = 202x - 249y $.

Таким образом, для того, чтобы это было выгодно компании, $202x$ должно быть больше, чем $249y$, или $x > 1.23y$.

Т.е. компания может взять на вооружение эту стратегию, если доход от абонента превысит расход более чем в 1.23 раз, или на 23%. Но при этом она лишится 133 абонентов.

Однако, можно пойти дальше и посмотреть все варианты сочетаний этих четырех метрик при разных порогах вероятности.

Для этого выведем общую формулу для сравнения доходов с расходами:

$ TPx - (TP + FP)y - FN(x - y) = TPx - TPy - FPy - FNx + FNy = (TP - FN)x - (TP + FP - FN)y $.

Т.о. $(TP - FN)x$ должно быть больше, чем $(TP + FP - FN)y$, или $ x > \frac{(TP + FP - FN)}{(TP - FN)}y $.

Упростив, получим: $$ x > (1 + \frac{FP}{(TP - FN)})y. $$

Построим график зависимости величины $ \frac{FP}{(TP - FN)} \times 100\% $ (т.е. на сколько процентов минимум должен доход с одного удержанного абонента превышать затраты на его удержание) от количества упущенных по недосмотру абонентов, т.е. от $FN$.

```python
# График зависимости (FP/(TP-FN))*100% (по оси ординат) от FN (по оси абсцисс)

x_lst = []
y_lst = []
y_max = 0
y_min = 0
for threshold in np.arange(0.0, 1.0, 0.005):
    predictions = probabilities[:, 1] > threshold
    matrix = confusion_matrix(target_test, predictions)
    x = matrix[1, 0]
    y = 100 * matrix[0, 1] / (matrix[1, 1] - matrix[1, 0])
    x_lst.append(x)
    y_lst.append(y)
    if y > y_max:
        y_max = y
    if y < y_min:
        y_min = y

plt.plot(x_lst, y_lst)
plt.xlabel('Количество упущенных абонентов')
plt.ylabel('Превышение доходов над расходами (%)')
plt.show()
```

Поскольку компания не может допустить, чтобы прибыль была отрицательной, рассмотрим подробнее только ту часть графика, где значения ординат положительные. Для этого определим в их списке индекс того значения, после которого график резко идет в минус - максимума функции. Его индекс в списке - это порядковый номер соответствующего порога вероятности. Умножив его на шаг 0.005, получим само значение порога, дальше которого идти нет смысла.

```python
thr = round(y_lst.index(y_max) * 0.005, 3)
print('Граница значений порога вероятности:', thr)
```

```python
# Обновляем границу порога и снова выводим график.

x_lst = []
y_lst = []
x_max = 100
x_min = 100
y_max = 100
y_min = 100

for threshold in np.arange(0.0, thr, 0.005):
    predictions = probabilities[:, 1] > threshold
    matrix = confusion_matrix(target_test, predictions)
    x = matrix[1, 0]
    y = 100 * matrix[0, 1] / (matrix[1, 1] - matrix[1, 0])
    x_lst.append(x)
    y_lst.append(y)
    if x > x_max:
        x_max = x
    if x < x_min:
        x_min = x
    if y > y_max:
        y_max = y
    if y < y_min:
        y_min = y
        
print("Максимальное количество абонентов:", x_max)
print("Минимальное количество абонентов:", x_min)
print(f"Максимальное превышение: {(y_max/100):.1%}")
print(f"Минимальное превышение: {(y_min/100):.1%}")

plt.figure(figsize=(9, 6))
plt.plot(x_lst, y_lst)
plt.xlabel('Количество упущенных абонентов')
plt.ylabel('Превышение доходов над расходами (%)')
plt.xticks(range(x_min, x_max, 20))
plt.yticks(np.arange(0, y_max, 25))
plt.show()
```

```python
# Находим количество упущенных абонентов при минимуме функции.

threshold_y_min = y_lst.index(y_min) * 0.005
predictions = probabilities[:, 1] > threshold_y_min
matrix = confusion_matrix(target_test, predictions)
x = matrix[1, 0]
print(f'Значение аргумента при минимальном значении функции ({(y_min/100):.1%}): {x}')
```

На этом графике в том числе можно найти точки, соответствующие тем четырем стратегиям, которые мы описали выше. Однако компания должна иметь возможность выбирать любую точку на нем в соответствии со своими потребностями и возможностями.

### Выводы и рекомендации

**1.** Поиск модели, дающей максимальное значение метрик, имеет важное значение. В частности, AUC-ROC, или площадь под ROC-кривой. Эта кривая наглядно показывает соотношение между полезным качеством - способностью правильно выявлять потенциально уходящих абонентов, и вредным - склонностью ошибочно считать уходящими тех, кто уходить не думал. Кривая образована множеством точек, координатами которых являются значения двух метрик: по оси ординат - TPR (доля истинно-положительных ответов от реально положительных значений, т.е. та же полнота, она же - "чувствительность"), а по оси абсцисс - FPR (доля ложно-положительных ответов от реально отрицательных значений, она же - единица минус "специфичность"). Если TPR - это доля абонентов, которых оператор догадался вовремя удержать, от всего их количества, кого нужно было удерживать, то FPR - это доля ошибочно выбранных абонентов от общего числа тех, кто не собирался никуда уходить. Иными словами, TPR - доля правильно понесенных расходов от всех необходимых расходов. А FPR - это доля зря понесенных затрат от всех потенциально ненужных затрат.


**2.** Смысл показателя AUC-ROC заключается в том, что он характеризует меру упорядоченности классов (единиц и нулей) при упорядоченности вероятностных оценок, т.е. насколько четко положительный класс будет отделен от отрицательного, если сделанные относительно них предсказания в виде вероятностей выстроить в порядке убывания (или возрастания). Иными словами, это более универсальная метрика и не зависит от порога вероятности, как остальные. Площадь под ROC-кривой состоит из сумм площадей множества четырехугольников, составленных из соседних точек на графике и соответствующих им точек на оси абсцисс, и математически этот показатель сводится к некоему абстрактному, усредненному значению полноты, с учетом всех возможный порогов классификации.


**3.** Таким образом, эта метрика демонстрирует общую способность модели правильно классифицировать. Она не показывает соотношение оправданных расходов ко всем расходам, понесенным в связи с прогнозом (как это делает *точность*), или же ко всем необходимым расходам (как это делает *полнота*). Между тем, оптимальное сочетание последних двух параметров имеет немалое значение для заказчика данного исследования. Точность для компании - это доля не зря потраченных ресурсов на предупреждение ухода клиентов: на их обзвон, на рекламную кампанию для их удержания, на скидки, различные акции и дополнительные бонусы. Все это важно, поскольку связано с расходом денег и времени. Полнота же - это доля вовремя удержанных клиентов, которые останутся и воспользуются новыми услугами, которые не уйдут к конкурентам. Это так же важно, поскольку связано с потенциальным недополучением прибыли.


**4.** Исследовательский анализ данных позволяет прийти к выводу, что, возможно, недавно как раз активизировались конкуренты оператора связи, предложившие абонентам компании дополнительные услуги, другие тарифные планы или иную периодичность платежей. Судя по результатам анализа, вполне вероятно отсутствие у компании должного внимания к потребностям своих клиентов, и в особенности - крупных. Не исключены случаи, когда абоненты забывали вовремя оплачивать счета, что влекло внезапное отключение услуг, а занятая линия техподдержки не позволяла им оперативно решить их проблемы.


**5.** Выведенная после анализа результатов обучения моделей кривая, показывающая зависимость минимально допустимой прибыли при удержании каждого абонента от количества упущенных абонентов, дает полезную для компании информацию. Она поможет ей принимать правильные решения после получения предсказаний модели. Зная примерно значение дохода с каждого потенциально удержанного абонента, с учетом себестоимости услуг, и заложив минимально приемлемую прибыль, финансовое управление компании может рассчитать оптимальный рекламный бюджет и выбрать нужную стратегию поведения.


**6.** Наконец, в качестве рекомендаций предлагаю обратить внимание компании на следующие моменты:
- побольше сосредоточиться на рекламе дополнительных услуг среди указанной группы абонентов;
- по возможности убеждать абонентов переходить на безакцептное списание с банковского счета или кредитной карты;
- отправлять счета на оплату одновременно разными способами (в бумажном виде, электронно, СМС-напоминаниями и пр.);
- при технической (случайной) просрочке оплаты не отключать связь сразу же, давать небольшое время на погашение долга;
- сделать техподдержку более доступной для абонентов, увеличить количество общих линий и снизить стоимость выделенных; 
- пересмотреть ценовую политику, изучив рынок, и создать новые, более конкурентные, тарифные планы; 
- добавить договоры с другой периодичностью платежей (раз в 3 месяца, в полгода) и предлагать их абонентам в качестве альтернативы;
- в части приобретения новых абонентов выбирать в качестве целевой аудитории прежде всего семейных людей с детьми.
